# **b) Stopword là gì?**

---

Stopwords là danh sách các từ không liên quan tới ý nghĩa của câu, dễ làm cho các công cụ tìm kiếm mơ hồ. Mỗi ngôn ngữ sẽ có một danh sách stop words khác nhau

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
##Hàm preprocess_text xử lý một đoạn text:
#1. bỏ tất cả các dấu câu
#2. đổi tất cả thành chữ thường
#3. bỏ tất cả các stopwords
#4. trả về list tất cả các từ (mỗi từ được địn nghĩa cách nhau bởi khoảng trắng)
import string
from nltk.corpus import stopwords
def preprocess_text(text):
    
    return [word for word in "".join([c for c in text.lower() if c not in string.punctuation]).split() if word not in stopwords.words('english')]
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
count_vectorizer = CountVectorizer(analyzer=preprocess_text).fit(text_train)

print(len(count_vectorizer.vocabulary_))
```

# **a) CountVectorizer là gì?**

---

Là vector chứa các từ đã được loại bỏ các stop word ra và các ký tự đặc biệt

```python
count_vectorizer.get_feature_names()
```

```python
X_train = count_vectorizer.transform(text_train)
X_test = count_vectorizer.transform(text_test)
```

```python
print(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(X_train, y_train)
```

# **c) Giải thích precision, recall và f1-score là gì?**

- Precision là tỉ lệ số điểm true positive trong số những điểm được phân loại là positive (TP + FP). Precision cao đồng nghĩa với việc độ chính xác của các điểm ta dự đoán là đúng là cao (khả năng dự đoán đúng nhãn).
- Recall được định nghĩa là tỉ lệ số điểm true positive trong số những điểm thực sự là positive (TP + FN). Recall cao đồng nghĩa với việc True Positive Rate cao, tức tỉ lệ bỏ sót các điểm thực sự positive là thấp (khả năng model dự đoán không bị sót nhãn)
- F1-score là harmonic mean (hàm trung bình điều hòa) của precision và recall. F1 càng cao, bộ phân lớp càng tốt. Khi cả recall và precision đều bằng 1 hay cả 2 cùng đều thấp 
`F1 = 2.(Precison + Recall)/(Precison*Recall)`

```python
from sklearn.metrics import classification_report
print(classification_report(y_train, model.predict(X_train),digits=4))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, model.predict(X_train),labels=['spam','ham'])
```

```python
print(classification_report(y_test, model.predict(X_test)))
```
